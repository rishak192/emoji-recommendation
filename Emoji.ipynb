```python
data=pd.read_csv('/content/drive/MyDrive/emoji/tweet_emotions.csv')
```

```python
# User handles
data['Clean_Text'] = data['content'].apply(nfx.remove_userhandles)
# Stopwords
data['Clean_Text'] = data['Clean_Text'].apply(nfx.remove_stopwords)
#Hashtags
data['Clean_Text'] = data['Clean_Text'].apply(nfx.remove_hashtags)
```

```python
data
```

```python
# Features & Labels
Xfeatures = data['Clean_Text']
ylabels = data['sentiment']
```

```python
#  Split Data
x_train,x_test,y_train,y_test = train_test_split(Xfeatures,ylabels,test_size=0.3,random_state=42)
```

```python
y_test=lb.transform(y_test)
y_train=lb.transform(y_train)
```

```python
tokenizer=Tokenizer(num_words=50000,oov_token="<OOV>")
tokenizer.fit_on_texts(Xfeatures)
word_index=tokenizer.word_index
print(word_index)
```

```python
len(word_index)
```

```python
x_train=tokenizer.texts_to_sequences(x_train)
x_test=tokenizer.texts_to_sequences(x_test)
```

```python
x_train=pad_sequences(x_train,padding='post')
x_test=pad_sequences(x_test,padding='post',maxlen=36)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
x_train[0]
```

```python
vs=33207
ed=64
model=tf.keras.Sequential([
    tf.keras.layers.Embedding(vs,50),
    # tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.LSTM(256,return_sequences=True),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(256,return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(256)),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(13,activation='softmax')
])
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
num_epochs=10
history=model.fit(x_train,y_train,epochs=num_epochs,validation_data=(x_test,y_test),verbose=1)
```
